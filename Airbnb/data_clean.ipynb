```python
import pandas as pd

listingDt = pd.read_csv('./listingDict.csv', names = 
 ['house_id',
 'title', 
'address',
'latitude', 
'longtitude', 
'checkin_rating',
'clean_rating',
'comm_rating',
'satis_overall',
'bathroom_type',
'bed_type',
'bedroom_type',
'guest_type',
'host_id',
'host_name'
])

listingDt = listingDt.loc[listingDt['address'] == 'College Park, Maryland, United States',]
listingDt['host_id'] = listingDt['host_id'].apply(lambda x: int(x))
listingDt.reset_index(drop=True,inplace=True)
listingDt['checkin_rating'].fillna(round(listingDt['checkin_rating'].mean(), 2), inplace=True)
listingDt['clean_rating'].fillna(round(listingDt['clean_rating'].mean(), 2), inplace=True)
listingDt['comm_rating'].fillna(round(listingDt['comm_rating'].mean(), 2), inplace=True)
listingDt['satis_overall'].fillna(round(listingDt['satis_overall'].mean(),2), inplace=True)
listingDt.drop_duplicates(inplace=True)
listingDt.to_csv('./cleaned/listingDt.csv', index = False)
listingDt
```

```python
hostDt = pd.read_csv('./hostDt.csv', names = 
    ['host_id',
    'first_name',
    'location',
    'join_date',
    'about',
    'identity_verified', 
    'picture_url'])

hostDt['host_id'] = hostDt['host_id'].apply(lambda x: int(x))
hostDt['identity_verified'] = hostDt['identity_verified'].apply(lambda x: int(x))
hostDt = hostDt[hostDt['host_id'].isin(listingDt['host_id'])]
hostDt.drop_duplicates(inplace=True)
hostDt.reset_index(drop=True,inplace=True)
hostDt.to_csv('./cleaned/hostDt.csv', index = False)
hostDt
```

```python
reviewDt = pd.read_csv('./reviewDt.csv')
reviewDt = reviewDt[reviewDt['house_id'].isin(listingDt['house_id'])]
reviewDt = reviewDt[reviewDt['reviewer_id'].isin(reviewerDt['reviewer_id'])]
reviewDt.drop_duplicates(inplace=True)
reviewDt.reset_index(drop=True,inplace=True)
reviewDt.to_csv('./cleaned/reviewDt.csv', index = False)
reviewDt
```

```python
reviewerDt = pd.read_csv('./reviewerDt.csv')

reviewerDt['identity_verified'] = reviewerDt['identity_verified'].apply(lambda x: int(x))
reviewerDt['reviewer_id'] = reviewerDt['reviewer_id'].apply(lambda x: int(x))
reviewerDt = reviewerDt[reviewerDt['reviewer_id'].isin(reviewDt['reviewer_id'])]
reviewerDt.drop_duplicates(inplace=True)
reviewerDt.reset_index(drop=True,inplace=True)
reviewerDt.to_csv('./cleaned/reviewerDt.csv', index = False)
reviewerDt
```

```python
priceDt = pd.read_csv('./priceDt.csv', names = ['Date','houseId','price'])
priceDt = priceDt[['houseId','Date','price']]
priceDt = priceDt[priceDt['houseId'].isin(listingDt['house_id'])]
priceDt.reset_index(drop=True,inplace=True)
priceDt.to_csv('./cleaned/priceDt.csv', index = False)
priceDt
```
